## Kütüphaneleri import edin

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

## Değişkenlerimizi tanımlayalım

m değerini output a göre numpy kütüphanesi yardımı ile oluşturalım

```python
m = np.arange(11,dtype = "float")
m
```

```python
c = 3* 10**8
```

```python
E = m * c**2
```

```python
E
```

## Oluşturduğumuz dokümanları plotlayalım

#### PLOTTING E=mc^2

```python
plt.figure(figsize = (6,4))
plt.plot(m,E,linewidth= 4, color = "b")
plt.xlabel("M in g")
plt.ylabel("E in j")
plt.title("E = mc^2");
plt.show()
```

```python
plt.figure(figsize = (6,4))
plt.plot(m,E,color = "orange")
plt.xlabel("M in g")
plt.ylabel("E in j")
plt.title("E = mc^2");
plt.show()
```

## Alttaki görseli oluştururken farklı bir değişken tanımlamayın. Bence eksenlerden birinin logaritmik hesaplanması gerekebilir:)

## Ayrıca axes de oluşan gridler bir style değil bence matplotlib library de gridleri araştırmanız gerekir

```python
m = np.arange(11,dtype = "float")
m
```

```python
c = 3* 10**8
```

```python
E = m * c**2
```

**"semilogy"** y eksenini logaritmik ölçekte çizerken, x eksenini normal ölçekte çizer

```python
plt.figure(figsize = (6,4))
plt.semilogy(m,E,color = "green")  
plt.xlabel("M in g")
plt.ylabel("E in j")
plt.title("E = mc^2")

plt.grid(which='both', axis="y")
plt.show()
```

## Aşağıdaki değişkenlere göre ilgili platlamaları yapın

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
plt.plot(labels,july16_2007, label = "july16_2007")
plt.plot(labels,july16_2020, label = "july16_2020")
plt.legend()
plt.show()
```

```python
plt.plot(labels,july16_2007, label = "july16_2007")
plt.plot(labels,july16_2020, label = "july16_2020")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

## Alttaki görselleri araştırarak yapmaya çalışalım. Zihnimizi ve parmaklarımızı zorlayalım:)

#### Bir hatırlatma "fig, ax = plt.subplots" syntax i size cevabı bulmanızda yardımcı olur. 

#### ayrıca shift + tab+ tab yapmayı unutmayın ve içindeki parametreleri denemeyi unutmayın.

```python
fig, ax = plt.subplots(2, 1, figsize=(10,6))

ax[0].plot(labels,july16_2007, label = "july16_2007")
ax[0].set_title("july16_2007")

ax[1].plot(labels,july16_2020, label = "july16_2020")
ax[1].set_title("july16_2020")
plt.tight_layout()                           # bu birbiri icine gecme durumunu, grifligi duzelten kod bu     
plt.show()
```

## Bu görselleri alabilenler için alkışı borç bilirim

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()     # bu işlem ile x eksenin ortak kullanımı sağlanır; ax1 ve ax2 y ekseninde farklı ölçekte

ax1.plot(labels, july16_2007, color='blue', label='July 16, 2007')
ax1.set_ylabel("july 16 2007",fontsize = 12, color = "b")


ax2.plot(labels, july16_2020, color='red', label='July 16, 2020')
ax2.set_ylabel('july 16 2020', color = "r",fontsize = 12)

ax2.spines['right'].set_color('red')
ax2.spines["left"].set_color("blue")

ax1.set_xlabel('Labels')
# ax1.set_title('Yield Curve Comparison: July 16, 2007 vs. July 16, 2020')  ax1 veya ax2 üzerinden ortak set_title çalışır
ax2.set_title('Yield Curve Comparison: July 16, 2007 vs. July 16, 2020')

ax1.legend(loc='upper left')
ax2.legend(loc='upper center')

plt.show()
```
