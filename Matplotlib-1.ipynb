```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## plt.plot

### lineer Ã§izimler

```python
x = [0, 1, 2, 3, 4, 5, 6]        #  x = np.arange(7)
y = [0, 2, 4, 6, 8, 10, 12]      #  y = np.arange(0,13,2)

plt.figure(figsize = (4, 2))     # grafik boyut ayarÄ±

plt.plot(x, y)

plt.show()
```

```python
x = np.arange(7)

y1 = -3 * x                          # f(x) = -3x grafiÄi
plt.figure(figsize = (4, 2))
plt.plot(x, y1, "--")                # plot biÃ§im ayarÄ±: renk,iÅaret --> "r-", "g^", "go", .......

plt.show()
```

```python
x = np.arange(7)

y2 = 3 * x + 5                                    # f(x) = 3x + 5   grafiÄi                                                                 
plt.figure(figsize = (4, 2))
plt.plot(x, y2, marker = "o", color = "red")      # marker ve color kullanÄ±mÄ±

plt.show()
```

### lineer olmayan Ã§izimler

```python
x = np.arange(100) / 10           # daha net sinÃ¼s formu yakalamak iÃ§in 10' bÃ¶ldÃ¼k (ayrÄ±ntÄ± yakalamak iÃ§in)
y = np.sin(x)                    

plt.plot(x,y)
```

```python
x = np.arange(10)
y1 = x * 2
y2 = x ** 2

plt.figure(figsize = (4, 2))
plt.plot(x,y1)          # 2 grafiÄi tek Ã§erÃ§evede gÃ¶sterme
plt.plot(x,y2)
```

```python
plt.figure(figsize = (4, 2))
plt.plot(x,y1, label = "x*2")               # etiket ekleme
plt.plot(x,y2, label = "x**2")
plt.legend()                                # plt.legend(loc="right"), (upper right) gibi etiketin konumunuda ayarlayabiliriz
```

## plt.scatter

```python
x = np.arange(7)
y = np.arange(0,13,2)

plt.figure(figsize = (4, 2))    

plt.scatter(x, y)

plt.show()
```

```python
plt.figure(figsize = (4, 2))    
plt.plot(x,y, color = "red")
plt.scatter(x, y, color = "blue")
```

```python
x = np.arange(100) / 10 
y1 = np.sin(x)
y2 = np.cos(x)

plt.scatter(x, y1, label = "sin", color = "red")
plt.plot(x, y2, label = "cos")
plt.legend(loc = "upper center")
plt.show()
```

parametreler:

`color`: Ã§izginin rengini belirler

`linestyle`: Ã§izgi stilini belirler

`marker`: noktalara denk gelen kesiÅimleri gÃ¶sterir. "o" olabilir, "s" olabilir.

`s`: noktalarÄ±n bÃ¼yÃ¼klÃ¼ÄÃ¼

`edgecolors`: noktalarÄ± saran renk.

`alpha`: saydamlÄ±k - opaklÄ±k. 1=opak, 0=saydam

`label`: grafiÄin etiketi

```python
plt.scatter(x, np.cos(x), color = "red", s = 40, label = "cos(x)", alpha = 0.5)
plt.legend()
plt.show()
```

```python
plt.scatter(x, np.sin(x), color = "red", s = 40 , edgecolors = "blue", linewidths = 10, label = "cos(x)", alpha = 0.5)
plt.legend()
plt.show()
```

```python
data = {
    "YÄ±l": np.arange(2010, 2021),
    "SatÄ±Å": np.random.randint(100, 500, 11),
    "Masraf": np.random.randint(50, 400, 11),
    "BÃ¶lge": np.random.choice(["Kuzey", "GÃ¼ney", "DoÄu", "BatÄ±"], size = 11)
}
df = pd.DataFrame(data)
df
```

```python
# yÄ±l-satÄ±Å ile yÄ±l-masraf plot grafiÄi aynÄ± tabloda

plt.plot(df["YÄ±l"],df["SatÄ±Å"], 
         label = "SatÄ±Ålar", 
         marker = "o", 
         color = "red", 
         linestyle = "--",
         markerfacecolor = "yellow")

plt.plot(df["YÄ±l"], df["Masraf"],
         label="Masraflar",
         linestyle = "--",
         marker="*", color="green")

plt.title("YÄ±llÄ±k SatÄ±Å ve Masraflar", fontsize = 18)
plt.xlabel("YÄ±llar", fontsize = 14)
plt.ylabel("TL", fontsize = 14)
plt.legend()
plt.show()
```

## plt.bar

`plt.bar()` bir sÃ¼tun grafiÄi Ã§izer. Bu grafik tÃ¼rÃ¼, kategorilerin deÄerleri arasÄ±ndaki iliÅkiyi gÃ¶sterir.

__Parametreler:__
    
`color`: SÃ¼tunlarÄ±n rengini belirtir.

`width`: SÃ¼tunlarÄ±n geniÅliÄini kontrol eder

`edgecolor`: SÃ¼tunlarÄ±n kenarlÄ±k rengini belirler

`linewidth`: KenarlÄ±k kalÄ±nlÄ±ÄÄ±nÄ± belirtir.

```python
categories = ["A", "B", "C", "D", "E"]
values = [35, 20, 45, 51, 10]

plt.bar(categories, values)
plt.show()
```

```python
plt.bar(categories, values, color = "red",
        width = 0.5, edgecolor="blue", linewidth=4)
plt.show()
```

## plt.hist

NÃ¼merik deÄerlerinizin daÄÄ±lÄ±mÄ±nÄ± yorumlamanÄ±zÄ± saÄlayan bir grafiktir, histogram Ã§izer.

__Parametreler__

`bins`: Grup sayÄ±sÄ±nÄ± belirler.

`range`: Bin'lerin alt ve Ã¼st limitini belirler.

`density`: True olursa grafik yoÄunluk belirtir, false olursa frekans belirtir.

`color`: Bin rengi

`alpha`: SaydamlÄ±k opaklÄ±k

```python
plt.hist(df["SatÄ±Å"], bins = 30, alpha = 0.5, color = "blue", label = "SatÄ±ÅlarÄ±n FrekansÄ±")
plt.legend()
plt.show()
```

```python
# numpy ile random veri oluÅturup plt.hist() ile frekansÄ±nÄ± gÃ¶rselleÅtirelim

plt.hist(np.random.randn(1000), bins = 30,
         color = "purple", label = "Frekans", width = 0.15)
plt.legend()
plt.show()
```

```python
plt.hist(np.random.randn(1000), bins = 30,
        alpha = 0.5, color = "purple", label = "Frekans", width = 0.15, range = (-5,5))    # range ile x ekseni daÄÄ±lÄ±mÄ± ayarlanÄ±r
plt.legend()
plt.show()
```

```python
plt.hist(np.random.randn(1000), bins = 30,
         color = "magenta", label = "Frekans", width = 0.15, density = True)   # density = True deÄerlerin yoÄunluÄunu gÃ¶sterir 
plt.legend()                                                                   # yukarÄ±dakinde deÄerlerin sayÄ±sÄ±nÄ± gÃ¶steriyordu
plt.show()
```
