```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## plt.plot

### lineer çizimler

```python
x = [0, 1, 2, 3, 4, 5, 6]        #  x = np.arange(7)
y = [0, 2, 4, 6, 8, 10, 12]      #  y = np.arange(0,13,2)

plt.figure(figsize = (4, 2))     # grafik boyut ayarı

plt.plot(x, y)

plt.show()
```

```python
x = np.arange(7)

y1 = -3 * x                          # f(x) = -3x grafiği
plt.figure(figsize = (4, 2))
plt.plot(x, y1, "--")                # plot biçim ayarı: renk,işaret --> "r-", "g^", "go", .......

plt.show()
```

```python
x = np.arange(7)

y2 = 3 * x + 5                                    # f(x) = 3x + 5   grafiği                                                                 
plt.figure(figsize = (4, 2))
plt.plot(x, y2, marker = "o", color = "red")      # marker ve color kullanımı

plt.show()
```

### lineer olmayan çizimler

```python
x = np.arange(100) / 10           # daha net sinüs formu yakalamak için 10' böldük (ayrıntı yakalamak için)
y = np.sin(x)                    

plt.plot(x,y)
```

```python
x = np.arange(10)
y1 = x * 2
y2 = x ** 2

plt.figure(figsize = (4, 2))
plt.plot(x,y1)          # 2 grafiği tek çerçevede gösterme
plt.plot(x,y2)
```

```python
plt.figure(figsize = (4, 2))
plt.plot(x,y1, label = "x*2")               # etiket ekleme
plt.plot(x,y2, label = "x**2")
plt.legend()                                # plt.legend(loc="right"), (upper right) gibi etiketin konumunuda ayarlayabiliriz
```

## plt.scatter

```python
x = np.arange(7)
y = np.arange(0,13,2)

plt.figure(figsize = (4, 2))    

plt.scatter(x, y)

plt.show()
```

```python
plt.figure(figsize = (4, 2))    
plt.plot(x,y, color = "red")
plt.scatter(x, y, color = "blue")
```

```python
x = np.arange(100) / 10 
y1 = np.sin(x)
y2 = np.cos(x)

plt.scatter(x, y1, label = "sin", color = "red")
plt.plot(x, y2, label = "cos")
plt.legend(loc = "upper center")
plt.show()
```

parametreler:

`color`: çizginin rengini belirler

`linestyle`: çizgi stilini belirler

`marker`: noktalara denk gelen kesişimleri gösterir. "o" olabilir, "s" olabilir.

`s`: noktaların büyüklüğü

`edgecolors`: noktaları saran renk.

`alpha`: saydamlık - opaklık. 1=opak, 0=saydam

`label`: grafiğin etiketi

```python
plt.scatter(x, np.cos(x), color = "red", s = 40, label = "cos(x)", alpha = 0.5)
plt.legend()
plt.show()
```

```python
plt.scatter(x, np.sin(x), color = "red", s = 40 , edgecolors = "blue", linewidths = 10, label = "cos(x)", alpha = 0.5)
plt.legend()
plt.show()
```

```python
data = {
    "Yıl": np.arange(2010, 2021),
    "Satış": np.random.randint(100, 500, 11),
    "Masraf": np.random.randint(50, 400, 11),
    "Bölge": np.random.choice(["Kuzey", "Güney", "Doğu", "Batı"], size = 11)
}
df = pd.DataFrame(data)
df
```

```python
# yıl ve satış plot grafiği

plt.plot(df["Yıl"],df["Satış"], 
         label = "Satışlar", 
         marker = "o", 
         color = "red", 
         linestyle = "--",
         markerfacecolor = "yellow")
plt.legend()
plt.show()
```

```python
# yıl-satış ile yıl-masraf plot grafiği aynı tabloda

plt.plot(df["Yıl"],df["Satış"], 
         label = "Satışlar", 
         marker = "o", 
         color = "red", 
         linestyle = "--",
         markerfacecolor = "yellow")

plt.plot(df["Yıl"], df["Masraf"],
         label="Masraflar",
         linestyle = "--",
         marker="*", color="green")

plt.title("Yıllık Satış ve Masraflar", fontsize = 18)
plt.xlabel("Yıllar", fontsize = 14)
plt.ylabel("TL", fontsize = 14)
plt.legend()
plt.show()
```

## plt.bar

`plt.bar()` bir sütun grafiği çizer. Bu grafik türü, kategorilerin değerleri arasındaki ilişkiyi gösterir.

__Parametreler:__
    
`color`: Sütunların rengini belirtir.

`width`: Sütunların genişliğini kontrol eder

`edgecolor`: Sütunların kenarlık rengini belirler

`linewidth`: Kenarlık kalınlığını belirtir.

```python
categories = ["A", "B", "C", "D", "E"]
values = [35, 20, 45, 51, 10]

plt.bar(categories, values)
plt.show()
```

```python
plt.bar(categories, values, color = "red",
        width = 0.5, edgecolor="blue", linewidth=4)
plt.show()
```

```python
data = {
    "Yıl": np.arange(2010, 2021),
    "Satış": np.random.randint(100, 500, 11),
    "Masraf": np.random.randint(50, 400, 11),
    "Bölge": np.random.choice(["Kuzey", "Güney", "Doğu", "Batı"], size = 11)
}
df = pd.DataFrame(data)
df
```

```python
Bölge_satıs = df.groupby("Bölge")["Satış"].sum()
Bölge_satıs
```

```python
type(Bölge_satıs)
```

```python
plt.bar(Bölge_satıs.index, Bölge_satıs.values)      # series olduğu için değerlere index ve values ile eriştik eğer çift parantez kullanıp
plt.title("Bölgelerin Toplam Satışı")               # dataframe çıktısı alsaydık bu şekilde çalıştıramazdık
plt.xlabel("Bölge") 
plt.ylabel("Satış")
plt.show()
```

```python
Bölge_satıs2 = df.groupby("Bölge")[["Satış"]].sum()
Bölge_satıs2
```

```python
type(Bölge_satıs2)
```

```python
plt.bar(Bölge_satıs2["Satış"].index, Bölge_satıs2["Satış"])         # dataframe üzerinden bar plot
plt.title("Bölgelerin Toplam Satışı")              
plt.xlabel("Bölge") 
plt.ylabel("Satış")
plt.show()
```

## plt.hist

Nümerik değerlerinizin dağılımını yorumlamanızı sağlayan bir grafiktir, histogram çizer.

__Parametreler__

`bins`: Grup sayısını belirler.

`range`: Bin'lerin alt ve üst limitini belirler.

`density`: True olursa grafik yoğunluk belirtir, false olursa frekans belirtir.

`color`: Bin rengi

`alpha`: Saydamlık opaklık

```python
plt.hist(df["Satış"], bins = 30, alpha = 0.5, color = "blue", label = "Satışların Frekansı")
plt.legend()
plt.show()
```

```python
# numpy ile random veri oluşturup plt.hist() ile frekansını görselleştirelim

plt.hist(np.random.randn(1000), bins = 30,
         color = "purple", label = "Frekans", width = 0.15)
plt.legend()
plt.show()
```

```python
plt.hist(np.random.randn(1000), bins = 30,
        alpha = 0.5, color = "purple", label = "Frekans", width = 0.15, range = (-5,5))    # range ile x ekseni dağılımı ayarlanır
plt.legend()
plt.show()
```

```python
plt.hist(np.random.randn(1000), bins = 30,
         color = "magenta", label = "Frekans", width = 0.15, density = True)   # density = True değerlerin yoğunluğunu gösterir 
plt.legend()                                                                   # yukarıdakinde değerlerin sayısını gösteriyordu
plt.show()
```
