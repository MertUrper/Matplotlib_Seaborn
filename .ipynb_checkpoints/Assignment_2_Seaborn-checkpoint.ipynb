## read data

if didn't work you can use this parameters: encoding="windows-1252"

```python
df = pd.read_csv("PoliceKillingsUS.csv", encoding = "windows-1252")   
df.head()                                                             
```

## Shape

```python
df.shape
```

## 1a: Box plot of age in both genders separeted by manner of death.

```python
df.manner_of_death.unique()   
```

#### INSTRUCTION: Demonstrate the Qurtiles of age in both genders separeted by manner of death.**

```python
df.dropna(inplace=True)
```

```python
plt.figure(figsize=(5, 4))

sns.boxplot(data=df, x="gender", y="age", hue="manner_of_death")
plt.xlabel('gender', fontsize=12)
plt.ylabel('age', fontsize=12)

plt.show()
```

## 1b: Swarm Plot 

#### INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
plt.figure(figsize=(5, 4))

sns.swarmplot(data=df, x="gender", y="age", hue="manner_of_death")
plt.xlabel('gender', fontsize=12)
plt.ylabel('age', fontsize=12)

plt.show()
```

## 2: Plot 12 most dangerous cities

```python
df["city"].value_counts().head(12)           
```

```python
# istenilen çıktı ile son 3 sırada bir farklılık oluştu anlamadım hocam!!!!
```

```python
df["city"].value_counts().head(12).values
```

```python
list(df["city"].value_counts().head(12).values)
```

```python
np.array(list(df["city"].value_counts().head(12).values))         # dtype=int64 kısmını kaldıramadım maalesef Hocam
```

```python
df["city"].value_counts().head(12).index
```

```python
plt.figure(figsize=(9, 4))

sns.barplot(data=df, x=df["city"].value_counts().head(12).index, y=df.city.value_counts().head(12).values)  

plt.title('Most Dangerous Cities', fontsize=10, color='r')
#plt.xticks(rotation=45)
plt.tick_params(axis='x', labelsize=7, rotation=45, colors='b')    # iki farklı şekilde rotation belirleme söz konusu
plt.tick_params(axis='y', labelsize=7)
plt.show()
```

```python
# with seaborn
```

```python
df["city"].value_counts().head(12)
```

```python
# aynı çıktıyı index ile de elde edebiliriz 
df["city"].value_counts()[:12]
```

```python
df["city"].value_counts().head(12).index
```

```python
plt.figure(figsize=(6, 4))

sns.countplot(data=df, x="city", order=df["city"].value_counts().head(12).index)    
                                                               
plt.title('Most dangerous cities', fontsize=10, color='b')
plt.tick_params(axis='x', labelsize=7, rotation=45)
plt.tick_params(axis='y', labelsize=7)
plt.xlabel("city", fontsize=7)
plt.ylabel("count", fontsize=7)
plt.show()
```

## 3: Plot 20 most dangerous states
