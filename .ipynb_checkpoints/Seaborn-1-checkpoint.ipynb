```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
plt.rcParams                               # garafik özelliklerinin default değerlerini gösterir
```

```python
# tüm grafiklerde geçerli olması için baştan tercih ettiğimiz değerleri atıyoruz
plt.rcParams["figure.figsize"] = (10,6)
plt.rcParams["xtick.labelsize"] = 14
plt.rcParams["ytick.labelsize"] = 14
plt.rcParams["axes.labelsize"] = 16
```

```python
data = sns.load_dataset("penguins")
```

```python
data.head()
```

### histplot

```python
sns.histplot(data=data, x = "flipper_length_mm")
```

```python
data_male = data[data["sex"] == "Male"]["flipper_length_mm"]
data_female = data[data["sex"] == "Female"]["flipper_length_mm"]

sns.histplot(data=data_female, label = "Female", alpha=0.5, color = "blue")
sns.histplot(data=data_male, label = "Male", alpha = 0.6, color = "orange")

plt.legend()
plt.show()
```

```python
# seaborn kolaylığı tek satırda
sns.histplot(data=data, x = "flipper_length_mm", hue = "sex")
plt.show()
```

### scatterplot

```python
sns.scatterplot(
    data=data,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
)
plt.title("bill_length_mm vs bill_depth_mm")
plt.show()
```

### barplot

```python
data.groupby(["species", "sex"])[["body_mass_g"]].mean().reset_index()
```

```python
grouped_data = data.groupby(["species", "sex"])["body_mass_g"].mean().reset_index()
sns.barplot(data = grouped_data,
            x="species", y="body_mass_g",
            hue="sex")
plt.show()
```

### subplots

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))          # fig, size genel standartlaşmış kullanımı 
axes
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))               

sns.scatterplot(
    data=data,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    ax=axes[0, 1]                     # grafiğin konum bilgisi
)

axes[0, 1].set_title("bill_length_mm vs bill_depth_mm")         # konum bilgisi üzerinden grafik başlığı
```

```python
fig, axes = plt.subplots(2,2, figsize=(12, 10))               

sns.scatterplot(
    data=data,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    ax=axes[0, 1]                     
)

sns.boxplot(
    data=data,
    x="species",
    y="body_mass_g",
    ax=axes[0, 0] 
)

sns.histplot(
    data = data,
    x="bill_length_mm",
    hue="species",
    ax = axes[1,0])

axes[0, 1].set_title("bill_length_mm vs bill_depth_mm") 
axes[0, 0].set_title("species vs body_mass_g")
```

### colormap

```python
grouped_data = data.groupby(["species", "sex"])["body_mass_g"].mean().reset_index()
grouped_data
```

```python
sns.barplot(data = grouped_data,
            x="species", y="body_mass_g")
plt.show()
```

```python
grouped_data = data.groupby(["species"])["body_mass_g"].mean()      # reset_index kullanmadan groupby yaptık
grouped_data
```

```python
sns.barplot(x = grouped_data.index,
            y = grouped_data.values)
plt.show()
```

```python
plt.get_cmap("viridis")
```

```python
cmap = plt.get_cmap("viridis")              # colormap seçip camp olarak atıyoruz
```

```python
norm = plt.Normalize(grouped_data.min(), grouped_data.max())     # datamızı 0-1 arasına normalize ettik
```

```python
colors = cmap(norm(grouped_data.values))                # datamızı renk aralığımıza göre uyarladık
colors
```

```python
fig, ax = plt.subplots(figsize=(12,9))          # 1'e 1 subplots default değeri 

sns.barplot(x = grouped_data.index,
            y = grouped_data.values,
            palette = colors)
ax.set_ylabel("Average body_mass_g")

cax = fig.add_axes([0.95,0.15,0.05,0.67])       # The dimensions [left, bottom, width, height] of the new Axes (deneyerek ayarlayabiliriz)
sm = plt.cm.ScalarMappable(cmap = cmap, norm = norm)    # cax, sm ve colorbar kısmı detay 
fig.colorbar(sm, cax = cax)
plt.show()
```
