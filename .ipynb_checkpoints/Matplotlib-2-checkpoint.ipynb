```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

## titanic data

```python
df = sns.load_dataset("titanic")
df.head()
```

### sex(male-female) / bar-pie 

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.bar(df.sex.value_counts().index, df.sex.value_counts()) 

plt.subplot(1,2,2)
plt.pie(df.sex.value_counts(), labels = df.sex.value_counts().index, autopct = "%1.2f%%")
plt.show()
```

### pclass / bar-pie

```python
plt.bar(df.pclass.value_counts().index, df.pclass.value_counts())  
plt.show()
plt.pie(df.pclass.value_counts(), labels = df.pclass.value_counts().index, autopct = "%1.2f%%")
plt.show()
```

### age / hist

```python
plt.figure(figsize=(12,6))
plt.hist(df.age,bins = 50)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

### pclass-survived / pie

```python
not_survived_class = df['pclass'][df['survived']==0]
survived_class = df['pclass'][df['survived']==1]

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.pie(not_survived_class.value_counts(), labels = not_survived_class.value_counts().index, autopct = "%1.2f%%")
plt.title('people who didn\'t survived')

plt.subplot(1,2,2)
plt.pie(survived_class.value_counts(), labels = survived_class.value_counts().index, autopct = "%1.2f%%")
plt.title('people who survived')
plt.show()
```

### fare-survived / hist

```python
not_survived_fare = df['fare'][df['survived']==0]
survived_fare = df['fare'][df['survived']==1]

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.hist(not_survived_fare)
plt.title("People who didn\'t survive")

plt.subplot(1,2,2)
plt.hist(survived_fare)
plt.title("People who survived")
plt.show()
```

### age - fare / scatter

```python
plt.scatter(df['age'], df['fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
```

### age - fare / boxplot

```python
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.boxplot(df['age'].dropna())
plt.ylabel('Age')
plt.title('Age Distribution')

plt.subplot(122)
plt.boxplot(df['fare'].dropna())
plt.ylabel('Fare')
plt.title('Fare Distribution')
plt.show()
```

```python
plt.boxplot([df['age'].dropna(), df['fare'].dropna()])      # tek bir grafikte
plt.xticks([1, 2], ['Age', 'Fare']) 
plt.ylabel('Value')
plt.title('Age and Fare Box Plot')
plt.show()
```

### age - sex / hist-pie

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
male = df[df['sex'] == 'male']['age']
female = df[df['sex'] == 'female']['age']

plt.hist(male, alpha=0.5, label='Male')
plt.hist(female, alpha=0.5, label='Female')

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age by Sex')
plt.legend()

plt.subplot(1,2,2)
plt.pie(df.sex.value_counts(), labels = df.sex.value_counts().index, autopct = "%1.2f%%")
plt.title('Distribution Sex')
plt.show()
```

## iris data

```python
df2 = sns.load_dataset("iris")
df2.head()
```

### sepal_length - sepal_width - species / scatter

```python
df2.species.value_counts()
```

```python
# 1. yol
plt.scatter(df2['sepal_length'][df2['species']=="setosa"],df2['sepal_width'][df2['species']=="setosa"])
plt.scatter(df2['sepal_length'][df2['species']=="versicolor"],df2['sepal_width'][df2['species']=="versicolor"])
plt.scatter(df2['sepal_length'][df2['species']=="virginica"],df2['sepal_width'][df2['species']=="virginica"])
plt.legend(["setosa","versicolor","virginica"])
plt.show()
```

```python
# 2. yol
species = df2['species'].unique()       

colors = ['blue', 'green', 'red']        # her bir çiçek türü için farklı bir renk 

for i in range(len(species)):
    plt.scatter(x=df2[df2['species']==species[i]]['sepal_length'], 
                y=df2[df2['species']==species[i]]['sepal_width'],
                c=colors[i], label=species[i])
    
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs. Sepal Width')
plt.legend()
plt.show()
```

### petal_length - petal_width - species / scatter

```python
plt.scatter(df2['petal_length'][df2['species']=="setosa"],df2['petal_width'][df2['species']=="setosa"])
plt.scatter(df2['petal_length'][df2['species']=="versicolor"],df2['petal_width'][df2['species']=="versicolor"])
plt.scatter(df2['petal_length'][df2['species']=="virginica"],df2['petal_width'][df2['species']=="virginica"])
plt.legend(["setosa","versicolor","virginica"])
plt.show()
```

### petal_length - petal_width - species / hist

```python
setosa = df2[df2['species'] == 'setosa']
versicolor = df2[df2['species'] == 'versicolor']
virginica = df2[df2['species'] == 'virginica']

plt.hist(setosa['petal_length'], alpha=0.5, label='setosa')
plt.hist(versicolor['petal_length'], alpha=0.5, label='versicolor')
plt.hist(virginica['petal_length'], alpha=0.5, label='virginica')

plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Length by Species')
plt.legend()
plt.show()
```

```python
setosa = df2[df2['species'] == 'setosa']
versicolor = df2[df2['species'] == 'versicolor']
virginica = df2[df2['species'] == 'virginica']

plt.hist(setosa['petal_width'], alpha=0.5, label='setosa')
plt.hist(versicolor['petal_width'], alpha=0.5, label='versicolor')
plt.hist(virginica['petal_width'], alpha=0.5, label='virginica')

plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Width by Species')
plt.legend()
plt.show()
```
